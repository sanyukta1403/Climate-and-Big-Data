```python
# time sequence t=seq(2015,2018,length=100) means 100 evenly spaced values from 2015 to 2018 

import numpy as np #numpy is library for math + arrays; np is a shorthand so i don't have to write numpy every time
import matplotlib.pyplot as plt #matplotlib is a plotting/graphing library; pyplot is the part that behaves like a simple graphing tool; plt is the shorthand

t = np.linspace(2015, 2018, 100) #define time sequence; linspace means inearly spaced numbers; it creates evenly spaced values between a start and end; start at 2015, end at 2018 and create 100 numbers evenly spread out

#define functions 
y1 = np.sin(2 * np.pi * (t - 0.1)) #matches first function y1=sin(2(t0.1))
y2 = np.cos(2 * np.pi * t) ** 2 #matches second function y2=cos2(2t); **2 means square it 

#plot both functions on the same figure 
plt.plot(t, y1, label='y = sin(2(t - 0.1))') #plots x axis (t) and y axis(y1); label gives it a name for the legend
plt.plot(t, y2, label='y = cos(2t)') #plots y2 on the same graph 

plt.xlabel('t (year)') #adds text under the x-axis
plt.ylabel('y') #adds text next to y-axis
plt.legend() #displays a key that matches each curve to its label
plt.title('Trigonometric Functions over Time') #adds a title to the graph 

plt.show() #presents graph on screen
```

# **Question 1.5**

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 2*np.pi, 100) #define x range; x has 100 values from 0 to 2
y = np.linspace(0, 2*np.pi, 100) #define y range; y has 100 values from 0 to 2

X, Y = np.meshgrid(x, y) #meshgrid creates (x,y) pairs to create a 2d grid

Z = (np.sin(X) ** 2) * (np.cos(Y - np.pi) ** 2) #define the function z 

plt.figure() #creates a new empty figure 
plt.contour(X, Y, Z) #draws contour lines showing where the function z has constant values
plt.xlabel('x') #label x axis as x 
plt.ylabel('y') #label y axis as y 
plt.title('Contour Plot of z = sin(x) cos(y  )') #add title describing the contour plot
plt.show() #display plot on screen 

plt.figure() #create new empty figure for colour contour plot 
plt.contourf(X, Y, Z) #fills region with colour 
plt.colorbar(label='z value') #adds a color scale showing how colors map to z-values
plt.xlabel('x') #label x axis as x 
plt.ylabel('y') #label y axis as y 
plt.title('Color Contour Plot of z = sin(x) cos(y  )') #add title 
plt.show() #display plot on screen 
```
